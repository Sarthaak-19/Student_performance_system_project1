```python
# importing library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
!pip install -q kaggle

!kaggle datasets download -d rabieelkharoua/students-performance-dataset

!unzip students-performance-dataset.zip

import pandas as pd

df = pd.read_csv("Student_performance_data _.csv")  # Use the filename created after unzip
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x=df[['StudyTimeWeekly']]
y=df['GPA']
```

```python
# model train
model=LinearRegression()
model.fit(x,y)
```

```python
predicted_score=model.predict(x)
```

```python
# valid regression matrix
mse=mean_squared_error(y,predicted_score)
mae=mean_absolute_error(y,predicted_score)
rmse=np.sqrt(mse)
r2=r2_score(y,predicted_score)
```

```python
print("MSE:",mse)
print("MAE:",mae)
print("RMSE:",rmse)
print("R2:",r2)
```

```python
# graph
plt.figure(figsize=(10,6))
plt.hist(df['GPA'],bins=30,edgecolor='k')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.title('Distribution of GPA')
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x,y,color='blue',label='Actual_score')
plt.plot(x,predicted_score,color='red',label='predicted_score (Regression_line)')
plt.legend()
plt.xlabel('Study Time')
plt.ylabel('GPA')
plt.title('Study Time vs GPA')
```
